### 3.4  We turn to the fascinating approach of complex-step differentiation. First, Taylor expand f (x+ih) around x, where f is an analytic function; take the imaginary part of both sides and divide by h to show that f ′(x) = Im[ f (x + ih)/h] +O(h2) holds.
```python
import numpy as np
import matplotlib.pyplot as plt

# Función f(x)
def f(x):
    """Función original f(x) = (1 - cos(x)) / x^2."""
    return (1 - np.cos(x)) / x**2 if x != 0 else 0

# Versión reducida de f(x)
def f_reducida(x):
    """
    Versión reducida de f(x) para evitar cancelación catastrófica.
    Utiliza la identidad trigonométrica: (1 - cos(x)) = 2 * sin^2(x / 2).
    """
    return (2 * (np.sin(x / 2)**2)) / x**2 if x != 0 else 0

# Segunda derivada analítica de f(x)
def analytical_second_derivative(x):
    """
    Calcula la segunda derivada analítica de f(x).
    Derivada calculada a partir de la expresión original.
    """
    if x == 0:
        return 0
    return (4 * (x * np.sin(x) + np.cos(x) - 1)) / x**4

# Aproximación de la segunda derivada usando diferencias centrales
def second_derivative_central(f, x, h):
    """
    Aproxima la segunda derivada de f en x usando diferencias centrales.
    """
    return (f(x + h) - 2 * f(x) + f(x - h)) / h**2

# Valores para evaluar
x_val = 0.004
h_values = [10**-1, 10**-2, 10**-6]  # Valores de h

# Segunda derivada analítica
f_double_prime_analytical = analytical_second_derivative(x_val)

# Evaluar con la versión original y reducida
approximations_original = [second_derivative_central(f, x_val, h) for h in h_values]
approximations_reduced = [second_derivative_central(f_reducida, x_val, h) for h in h_values]

# Calcular errores
errors_original = [abs(f_double_prime_analytical - approx) for approx in approximations_original]
errors_reduced = [abs(f_double_prime_analytical - approx) for approx in approximations_reduced]

# Imprimir resultados detallados
print("Resultados:")
print("Analítico:", f_double_prime_analytical)
print("\nAproximaciones Originales y Errores:")
for h, approx, error in zip(h_values, approximations_original, errors_original):
    print(f"h = {h:.1e}, Aproximación = {approx:.6e}, Error = {error:.6e}")

print("\nAproximaciones Reducidas y Errores:")
for h, approx, error in zip(h_values, approximations_reduced, errors_reduced):
    print(f"h = {h:.1e}, Aproximación = {approx:.6e}, Error = {error:.6e}")

# Gráfica comparativa
plt.figure(figsize=(10, 6))
plt.loglog(h_values, errors_original, marker='o', label="Error absoluto (Original)", linestyle='--')
plt.loglog(h_values, errors_reduced, marker='s', label="Error absoluto (Reducida)", linestyle='-')
plt.title("Comparación del error absoluto en la aproximación de la segunda derivada")
plt.xlabel("Tamaño de paso (h)")
plt.ylabel("Error absoluto")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def phi(x, y):
    return 2*x**2 + 2*y**2

def gradiente(phi, x, y, h):
    dphi_dx = (phi(x + h, y) - phi(x, y)) / h
    dphi_dy = (phi(x, y + h) - phi(x, y)) / h
    return dphi_dx, dphi_dy
 
def Gradiente_r2(phi, x_range, y_range, h):
    x_values = np.linspace(x_range[0], x_range[1], 100)
    y_values = np.linspace(y_range[0], y_range[1], 100)
    X, Y = np.meshgrid(x_values, y_values)
    
    grad_x = np.zeros_like(X)
    grad_y = np.zeros_like(Y)
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            grad_x[i, j], grad_y[i, j] = gradiente(phi, X[i, j], Y[i, j], h)
    print(X.shape[0], X.shape[1])
    return X, Y, grad_x, grad_y

h = 1e-10
x_range = (-1, 1)
y_range = (-1, 1)


X, Y, grad_x, grad_y = Gradiente_r2(phi, x_range, y_range, h)

plt.figure(figsize=(12, 6))

plt.contourf(X, Y, grad_x*grad_y, levels=10, cmap='inferno')
plt.colorbar(label='∂φ/∂x*∂φ/∂y')
plt.title('∂φ/∂x +  ∂φ/∂y')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```

## 3.11

We turn to the relationship between Horner's rule, Eq. (2.100), and the evaluation of the first derivative of a polynomial. To bridge the two, introduce synthetic division:

$$
p(x) = (x - x_0)q(x) + p(x_0) 
$$

where $p(x)$ is of degree $n - 1$ and $q(x)$ of degree $n - 2$. Show that $p'(x_0) = q(x_0)$. Then, figure out the pattern obeyed by the coefficients of the different monomials in $q(x)$; use it to implement a version of Horner's rule that also computes $p'(x_0)$.


```python
#Definimos una función Horner con los (coeffs) coeficientes del polinomio de mayor a menor grado, (x0) punto que se va a 
# evaluar el polinomio y su derivada
def d_horner(coeffs, x0):
    #Grado del polinomio
    n = len(coeffs) - 1 
    # p(x0) inicializado de mayor a menor grado
    p = coeffs[0] 
    # dp(x0) empieza en cero
    dp = 0  
    #Creamos un for de 1 a n+1
    for i in range(1, n + 1):
        #Actualiza a nuevo dp
        dp = dp * x0 + p  
        #Actualiza a p(x0)
        p = p * x0 + coeffs[i]
    # Me regresa p(x0) y dp(x0)
    return p, dp

# Ejemplo
coeffs = [1, -3, 2, -1]  # Representa el polinomio p(x) = x^3 - 3x^2 + 2x - 1
x0 = 2
p_x0, dp_x0 = d_horner(coeffs, x0)

print(f"p({x0}) = {p_x0}")
print(f"p'({x0}) = {dp_x0}")
```
